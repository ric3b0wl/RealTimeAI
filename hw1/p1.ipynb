```python
from google.colab import drive
drive.mount('/content/drive')

# from skimage.transform import resize
# from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import numpy as np
import torch
import imageio


device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

red_dir = "/content/drive/My Drive/ECGR4106-RealTimeAI/hw1/imgs/red.jpg"
green_dir = "/content/drive/My Drive/ECGR4106-RealTimeAI/hw1/imgs/green.jpg"
blue_dir = "/content/drive/My Drive/ECGR4106-RealTimeAI/hw1/imgs/blue.jpg"
```

```python
red_img = imageio.imread(red_dir)
print("NumPy: ", type(red_img))
print("Shape: ", red_img.shape)
print("dtype: ", red_img.dtype)

red_img_np = torch.from_numpy(red_img)
print('Mean value Red: ', torch.mean(red_img_np.float()) )

# print(red_img)
```

```python
green_img = imageio.imread(green_dir)
print("NumPy: ", type(green_img))
print("Shape: ", green_img.shape)
print("dtype: ", green_img.dtype)

green_img_np = torch.from_numpy(green_img)
print('Mean value Green: ', torch.mean(green_img_np.float()) )

# print(green_img)
```

```python
blue_img = imageio.imread(blue_dir)
print("NumPy: ", type(blue_img))
print("Shape: ", blue_img.shape)
print("dtype: ", blue_img.dtype)

blue_img_np = torch.from_numpy(blue_img)
print('Mean value Blue: ', torch.mean(blue_img_np.float()) )

# print(blue_img)
```
